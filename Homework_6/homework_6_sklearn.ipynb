Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
boston = load_boston()
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
X.head()
```

```python
y = pd.DataFrame(boston.target, columns=["price"])
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# Обучение модели на тренировочных данных
lr.fit(X_train, y_train)
# Предсказание по тестовым данным
y_pred = lr.predict(X_test)
```

```python
# Вычислим метрику R2
from sklearn.metrics import r2_score

lr_r2 = r2_score(y_pred, y_test)
lr_r2
```

```python
# R2 - коэффициент детерминации, возможные значения от 0 до 1. Чем ближе к 1, тем лучше. 
# В нашем случае получилось 0.6693702691495615 - достаточно далеко от 1, с помощью линейной регресси результат получился
# не очень хороший. Стоит посмотреть другие модели.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
import pandas as pd

df = pd.read_csv('creditcard.csv', sep=',')
df.head()
```

```python
# df["column_name"].value_counts(normalize=True) - возвращает относительную частоту уникальных значений.
# Т.е. в долях относительно остальных значений по всему признаку
# Посмотрим, как распределяются значения в нашем целевом признаке Class - принадлежность операции к классу мошеннических (1), 
# или обычная операция (0)
df1 = df["Class"].value_counts(normalize=True)
df1
```

```python
# Видим, что примерно 99,8% - это обычные операции. И только 0,17% - это мошеннические операции. 
# Т.е. данные несбалансированы. Слишком много информации об обычных операциях и слишком мало - о мошеннических.
```

```python
# Проверим типы данных, нет ли пропусков
df.info()
```

```python
# Все данные заполнены, пропусков нет. Все значения числовые. 
```

```python
pd.options.display.max_columns = 100
df.head(10)
```

```python
X = df.drop("Class", axis=1)
X.head()
```

```python
y = pd.Series(df["Class"])
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
```

```python
# Зададим параметры для поиска по сетке (для подбора наилучших параметров)
parameters = [{'n_estimators': [10, 15],
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]
```

```python
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100), # в качестве модели RandomForest-классификатор
                     param_grid=parameters, # параметры для перебора, среди которых будут найдены оптимальные
                     scoring='roc_auc', # метрика оценки качества
                     cv=3) # кол-во моделей
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)
```

```python
y_pred_proba = y_pred_proba[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
